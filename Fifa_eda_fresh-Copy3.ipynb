```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
fifa=pd.read_csv('fifa18complete.csv')
```

```python
fifa = fifa[['Name', 'Age', 'Nationality', 'Overall', 'Wage', 'Potential', 'Club', 'Value', 'Preferred Positions']]
fifa.head(10)
```

```python

#For simplicity, we will only take the Data in which we are interested in:
fifa = fifa[['Name','Age','Nationality','Preferred Positions','Overall','Potential','Club','Value']]
fifa.head(10)
```

```python
#Now we need Values such as "Remaining Potential" , "First Preferred Position" and converting the "K" to "M" in the Value column.
# 1. 
fifa['Remaining Potential'] = fifa['Potential'] - fifa['Overall']
# 2.
fifa['1st Pref Position'] = fifa['Preferred Positions'].str.split().str[0]
# 3.
fifa['Unit'] = fifa['Value'].str[-1]
fifa['Value (M)'] = np.where(fifa['Unit'] == '0', 0, fifa['Value'].str[1:-1].replace(r'[a-zA-Z]',''))
fifa['Value (M)'] = fifa['Value (M)'].astype(float)
fifa['Value (M)'] = np.where(fifa['Unit'] == 'M', fifa['Value (M)'], fifa['Value (M)']/1000)
fifa = fifa.drop('Unit', 1)

fifa.head(20)
```

### Let's clean wage and value first. Using the function from this kernel.

```python
def extract_value_from(value):
    out = value.replace('€', '')
    if 'M' in out:
        out = float(out.replace('M', ''))*1000000
    elif 'K' in value:
        out = float(out.replace('K', ''))*1000
    return float(out)
```

```python
fifa['Value'] = fifa['Value'].apply(lambda x: extract_value_from(x))
fifa['Wage'] = fifa['Wage'].apply(lambda x: extract_value_from(x))
```

### Let's explore in detail the situation for the top teams

```python
grouped = fifa.groupby('Club')
wages = grouped.sum()['Wage'].sort_values(ascending = False)

#top 30 club wage-wise
top30 = wages.iloc[0:30].index
fifa_top30 = fifa[fifa['Club'].isin(top30)]

ax = sns.boxplot(x = 'Club', 
                 y = 'Wage', 
                 data = fifa_top30, 
                 order = top30,
                 linewidth = 1
                )

ax.set_xticklabels(labels = top30, rotation='vertical')
ax.set_ylabel('Wage (€)')
ax.set_xlabel('Club')
ax.set_title('Salary distribution per club')
```

#### in the last plot, the order on the x-axis is determined by the sum of the wages of the players in each club,, in decrescent order. On the other hand, the values presented in the boxplot, relate to the mean and to the quartiles. Therefore, we can observe some 'inversions'. For instance, we see that FC Barcelona has a lower mean wage if compared to Real Madrid CF , and the whiskers are less extended; however, there are more outliers in the former team than in the latter. Also, it is possible that the total number of players in each team is not the same, resulting in a discrepancy between the order of the mean and of the sum.

```python
def transform_compound(value):
    out = str(value)
    if '+' in out:
        a,b  = out.split('+')
        out = int(a)+int(b)
    elif '-' in out:
        a,b = out.split('-')
        if len(a)== 0: out = -1
        else: out = int(a)-int(b)
    elif out == '':
        out = 0
    return float(out)
```

### playing squad on the basis of position  and overall basis

```python
#Alright........Now let's have a look at the All Time Best XI squad that can be formed using this Data Set with each position's best Player:
def get_best_eleven(position):
    fifa_copy = fifa.copy()
    squad = []
    for i in position:
        squad.append([i,fifa_copy.loc[[fifa_copy[fifa_copy['fifa_new_ST'] == i]['Overall'].idxmax()]]
                      ['Name'].to_string(index = False), fifa_copy[fifa_copy['fifa_new_ST'] == i]['Overall'].max()])
        fifa_copy.drop(fifa_copy[fifa_copy['fifa_new_ST'] == i]['Overall'].idxmax(),inplace=True)
    return pd.DataFrame(np.array(squad).reshape(11,3), columns = ['Position', 'Player', 'Overall']).to_string(index = False)
    
positions_4_3_3 = ['GK','LB','CB','CB','RB','LM','CDM','RM','LW','ST','RW']
print('Best "4--3--3" Squad Ever..!!')
print(get_best_eleven(positions_4_3_3))
```
